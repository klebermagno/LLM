```python
# -*- coding: utf-8 -*-

from transformers import SpeechT5HifiGan, SpeechT5Processor, SpeechT5ForTextToSpeech
from datasets import load_dataset

import soundfile as sf
import torch
from IPython.display import Audio
import IPython

processor = SpeechT5Processor.from_pretrained("microsoft/speecht5_tts")
model = SpeechT5ForTextToSpeech.from_pretrained("microsoft/speecht5_tts")
vocoder = SpeechT5HifiGan.from_pretrained("microsoft/speecht5_hifigan")
embeddings_dataset = load_dataset("Matthijs/cmu-arctic-xvectors", split="validation")

```

```python
mytext="""
Hi, everyone! Today, we're diving into an essential concept in system design the CAP theorem. It's a fundamental principle that guides the design of distributed systems.

It's consist that we can only have two of the following three guarantees achieved simultaneously: Consistency, Availability, and Partition Tolerance.
"""
inputs = processor(text=mytext, return_tensors="pt")
```

```python

# load xvector containing speaker's voice characteristics from a dataset
speaker_embeddings = torch.tensor(embeddings_dataset[7306]["xvector"]).unsqueeze(0)

speech = model.generate_speech(inputs["input_ids"], speaker_embeddings, vocoder=vocoder)

sf.write("hgWells.wav", speech.numpy(), samplerate=16000)


out_audio = "hgWells.wav"
IPython.display.Audio(out_audio)
```
